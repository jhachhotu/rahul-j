```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Load the dataset
data = pd.read_csv('StudentsPerformance.csv')

# Display the first few rows of the dataset
print(data.head())
```

```python
# Example DataFrame
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Hours': [2.5, 5.1, 3.2, 8.5, 3.5],
    'Scores': [21, 47, 27, 75, 30]
})

# Print column names to verify
print(data.columns)  # Should output: Index(['Hours', 'Scores'], dtype='object')

# Create a scatter plot
plt.scatter(data['Hours'], data['Scores'])

# Add a title to the plot
plt.title('Hours vs Percentage')

# Label the x-axis
plt.xlabel('Hours Studied')

# Label the y-axis
plt.ylabel('Percentage Score')

# Display the plot
plt.show()
```

```python
# Split the data into features and target variable
X = data['Hours'].values.reshape(-1, 1)
Y = data['Scores'].values

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Create the linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, Y_train)
```

```python
# Make predictions on the test set
Y_pred = model.predict(X_test)
```

```python
# Evaluate the model's performance
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Plot the regression line
plt.scatter(X, Y, color='blue')  # Plot the actual data
plt.plot(X, model.predict(X), color='red')  # Plot the regression line
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```
